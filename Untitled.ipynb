```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/home/aman/Desktop/Project/Customer Support ChatBot/Mental_Health_FAQ.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
import re

def correct_spelling(text):
    return str(TextBlob(text).correct())

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)

    tokens = word_tokenize(text)

    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words
    ]

    return " ".join(tokens)

df["processed_question"] = df["Questions"].apply(preprocess)
```

```python
print(df[["Questions", "processed_question"]].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["processed_question"])

print(X.shape)
```

```python
print(vectorizer.get_feature_names_out()[:20])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def chatbot(user_query):

    user_query = correct_spelling(user_query)

    query_processed = preprocess(user_query)

    query_vector = vectorizer.transform([query_processed])

    similarities = cosine_similarity(query_vector, X)

    best_match = similarities.argmax()
    score = similarities[0][best_match]

    print("Matched Question:")
    print(df.iloc[best_match]["Questions"])
    print(f"Similarity Score: {score:.3f}")

    if score < 0.3:
        return "Sorry, I don't understand that question."

    return df.iloc[best_match]["Answers"]
```

```python
print(chatbot("What are symptoms of mental illness?"))
```

```python
while True:
    user = input("You: ")

    if user.lower() == "exit":
        break

    print("Bot:", chatbot(user))
```
